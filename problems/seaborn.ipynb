```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#create dataset
dataset = np.random.randn(25)

#create rugplot
sns.rugplot(dataset)

#set x axis or for the plot
x_min = dataset.min() - 2
x_max = dataset.max() + 2

# 100 equally spaced points
x_axis = np.linspace(x_min, x_max, 100)

#https://en.wikipedia.org/wiki/Kernel_density_estimation#Practical_estimation_of_the_bandwidth

bandwidth = ((4 * dataset.std() ** 5)/(3 * len(dataset)))**.2

#empty kernal list
kernal_list = []

for data_point in dataset :
    kernal = stats.norm(data_point, bandwidth).pdf(x_axis)
    kernal_list.append(kernal)
    
    kernal = kernal / kernal.max()
    kernal = kernal * .4
    plt.plot(x_axis, kernal, color = 'grey', alpha = 0.5)
    
plt.ylim(0, 1)
```

```python
#plot sum of basis function
sum_of_kde = np.sum(kernal_list, axis = 0)

fig = plt.plot(x_axis, sum_of_kde, color = 'indianred')

#add regplot
sns.rugplot(dataset, color='indianred')

plt.yticks([])
plt.suptitle('Sum of basis functions')
```
