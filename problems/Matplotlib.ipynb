```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y, 'r-' )
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title String')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
#Object oriented
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Outer plot')
axes2.plot(y, x)
axes2.set_title('Inner plot')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
#axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()

#for current_ax in axes:
#    current_ax.plot(x, y)

axes[0].plot(x, y)
axes[0].set_title('First')
axes[1].plot(y, x)
axes[1].set_title('Second')
```

```python
#fig = plt.figure(figsize=(3,2), dpi=100)
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure(figsize=(8, 2))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(8, 4))

axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X Square')
ax.plot(x, x**3, label='Y Cube')

ax.legend(loc=0)
```

# Plot Style

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, color='orange', label='X Square', linewidth=5, alpha=0.5, marker='.', markersize=20) # alos can add rgb hex code
ax.plot(x, x**3, color='blue', label='Y Cube', linewidth=3, linestyle='-.') # ls or lw
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='orange', lw=2, ls='--')
ax.set_xlim([0, 2])
ax.set_ylim([0, 5])
```
